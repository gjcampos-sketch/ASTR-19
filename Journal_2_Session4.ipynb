```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lam_val = 10.0
random_variates = np.random.poisson(lam=lam_val, size=1000)
```

```python
plt.figure(figsize=(7, 5))
plt.hist(random_variates, bins=100, facecolor='cornflowerblue', edgecolor='black', alpha=0.7)

plt.xlabel('Value of Random Variation')
plt.ylabel('Number of Counts')
plt.title(f'Histogram of 1000 Poisson Random Numbers ($\lambda={lam_val}$)')

plt.savefig('random_distribution_histogram.pdf', format='pdf', bbox_inches='tight')

plt.show()
```
