```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def main():
    x = np.linspace(0, 10, 50)
    y_model = 2 * x + 1

    sigma = np.linspace(2, 0.2, len(x))
    
    scatter = np.random.normal(scale=sigma)
    y_data = y_model + scatter
    
    plt.figure(figsize=(7, 5))
    
    plt.errorbar(x, y_data, yerr=sigma, fmt='^', color='cornflowerblue', 
                 ecolor='gray', elinewidth=1, capsize=2, label='Measured Data')
    
    plt.xlabel('X Coordinate')
    plt.ylabel('Y (Model + Variable Scatter)')
    plt.legend()
    
    plt.savefig('variable_scatter_plot.pdf', format='pdf', bbox_inches='tight')
    plt.show()

   
if __name__ == "__main__":
    main()
```
