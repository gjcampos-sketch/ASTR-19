```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2 * np.pi, 100)
def exponential_model(x, a, b):
    return a * np.exp(-b * x)

a_value = 1.0
b_value = 0.5
```

```python
y = exponential_model(x, a_value, b_value)
plt.plot(x, y, label=f'a={a_value}, b={b_value}')

plt.xlabel('x')
plt.ylabel('y = a * exp(-b * x)')

plt.savefig('exponential_plot.pdf', bbox_inches='tight', format='pdf')

plt.show()
```
